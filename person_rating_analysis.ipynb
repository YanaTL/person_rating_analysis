```python
import pandas as pd
people_df = pd.read_csv('applications(2.0).csv')

people_df
```

```python
del_duplicates = people_df.drop_duplicates(subset='applicant_id', keep='first')
del_duplicates.loc[:, 'External Rating'] = del_duplicates['External Rating'].fillna(0)
del_duplicates.loc[:, 'Education level'] = del_duplicates['Education level'].fillna('Середня')

print ("Очищення таблиці - Дз.Пункт 1")
del_duplicates
```

```python
import pandas as pd
industry_df = pd.read_csv('industries(2.0).csv')

industry_df
```

```python
two_table_ind = pd.merge(del_duplicates, industry_df, on='Industry', how='left')

print("Об'єднання двох таблиць - Дз.Пункт 2")
two_table_ind
```

```python
import math

def rating_calc(df):
    if math.isnan(df["Amount"]) or df["External Rating"] == 0:
        return 0
    rating = 0
    if df["Age"] >= 35 and df["Age"] <= 55:
        rating += 20
    appliedAt = pd.to_datetime(df["Applied at"])
    if appliedAt.weekday() < 5:
        rating += 20
    if df["Marital status"] == "Married":
        rating += 20
    if df["Location"] == "Київ чи область":
        rating += 10
    rating = df["Score"] + rating
    if df["External Rating"] >= 7:
        rating += 20
    if df["External Rating"] <= 2:
        rating -= 20
    return rating
    

two_table_ind['Rating'] = two_table_ind.apply(lambda x: rating_calc(x), axis=1)
new_filt_table = two_table_ind[(two_table_ind["Rating"] > 0) & (two_table_ind["Rating"] <= 100)]

print("Розрахований рейтинг за критеріями - Дз.Пункт 3")
new_filt_table
```

```python
filtr_date = new_filt_table[["Applied at", "Rating"]].copy()
filtr_date.loc[:,"Datetime"] = pd.to_datetime(new_filt_table["Applied at"], format = "mixed")
filtr_date.loc[:, "Week"] = filtr_date["Datetime"].dt.isocalendar().week
grouped_week_rating = filtr_date.groupby("Week")["Rating"].mean().reset_index()

print ("Згрупований середній рейтинг за тижнем подачі заявки - Дз.Пункт 5")
grouped_week_rating
```
